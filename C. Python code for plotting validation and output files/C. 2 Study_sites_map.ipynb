Plot Map with Study Sites

Guyaflux (Paracou): latitude = 5.25, longitude = -52.75
Santarem (Tapajos): latitude = -2.75, longitude = -54.75


```python
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap

# Create a Basemap instance
m = Basemap(projection='merc', llcrnrlat=-10, urcrnrlat=10, llcrnrlon=-60, urcrnrlon=-40, resolution='i')

# Paracou coordinates
paracou_lat = 5.25
paracou_lon = -52.75

# Tapajos coordinates
tapajos_lat = -2.75
tapajos_lon = -54.75

# Convert coordinates to map projection
paracou_x, paracou_y = m(paracou_lon, paracou_lat)
tapajos_x, tapajos_y = m(tapajos_lon, tapajos_lat)

# Create the map
m.drawcoastlines()
m.drawcountries()
m.drawmapboundary(fill_color='lightblue')
m.fillcontinents(color='white', lake_color='lightblue')
m.drawparallels(range(-90, 91, 4), labels=[1, 0, 0, 0], linewidth=0.2)
m.drawmeridians(range(-180, 181, 4), labels=[0, 0, 0, 1], linewidth=0.2)

# Plot the red dots with different marker styles
m.plot(paracou_x, paracou_y, 'r^', markersize=8, label='Paracou')
m.plot(tapajos_x, tapajos_y, 'ro', markersize=8, label='Tapajos')

# Add legend
plt.legend()

# Add title
plt.title('Study Sites')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap

# Create a Basemap instance
m = Basemap(projection='merc', llcrnrlat=-30, urcrnrlat=10, llcrnrlon=-80, urcrnrlon=-30, resolution='i')

# Paracou coordinates
paracou_lat = 5.25
paracou_lon = -52.75

# Tapajos coordinates
tapajos_lat = -2.75
tapajos_lon = -54.75

# Convert coordinates to map projection
paracou_x, paracou_y = m(paracou_lon, paracou_lat)
tapajos_x, tapajos_y = m(tapajos_lon, tapajos_lat)

# Create the map
m.drawcoastlines()
m.drawcountries()
m.drawmapboundary(fill_color='lightblue')
m.fillcontinents(color='white', lake_color='lightblue')
m.drawparallels(range(-30, 11, 6), labels=[1, 0, 0, 0], linewidth=0.2)  # Change latitude intervals
m.drawmeridians(range(-80, -29, 6), labels=[0, 0, 0, 1], linewidth=0.2)  # Change longitude intervals

# Plot the red dots with different marker styles
m.plot(paracou_x, paracou_y, 'r^', markersize=9, label='Paracou')
m.plot(tapajos_x, tapajos_y, 'ro', markersize=9, label='Tapajos')

# Add legend
plt.legend()

# Add title
plt.title('Study Sites')

# Show the plot
plt.show()
```
