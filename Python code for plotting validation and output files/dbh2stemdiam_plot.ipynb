```python
import numpy as np
import matplotlib.pyplot as plt

fig, axs = plt.subplots(1, 1, figsize=(5, 4))


# Declare allometry constants
allom1 = 250
allom2 = 40
allom3 = 0.67

# Define function to compute dbh from height
def dbh_from_height(height):
    stemdiam = (height / allom2) ** (1.0 / allom3)
    # Uncomment the line below and adjust the unit in the legend if you want to convert stemdiam from meters to centimeters
    # stemdiam = stemdiam * 100
    return stemdiam

# Generate a range of height values
heights = np.linspace(5, 100, 100)

# Calculate the corresponding stem diameters (dbh)
dbhs = dbh_from_height(heights)

# Plot the function
plt.plot(heights, dbhs)
plt.xlabel('Height (m)')
plt.ylabel('Stem Diameter (m)')
plt.title('Stem diameter as a function of height')
# plt.legend()

# Show the plot
plt.show()
```
